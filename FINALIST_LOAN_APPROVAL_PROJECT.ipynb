#**Project Introduction**

In today’s modern banking system, loan approval is a very important financial process. Banks and financial institutions receive thousands of loan applications, and manually checking each application takes a lot of time and effort. Manual decision-making can also lead to errors or inconsistencies. Therefore, there is a need for an automated and intelligent system that can help in faster and more accurate decision-making.

This project focuses on building a Loan Approval Prediction System using Machine Learning techniques. The main objective of this system is to predict whether a loan application will be approved or rejected based on the applicant’s financial and personal information.

The system analyzes important factors such as:

* Applicant’s age and occupation status

* Employment experience and annual income

* Credit score and credit history

* Savings and current debt

* Loan amount and interest rate

* Debt-to-income and payment-to-income ratios

By studying these features, the model identifies patterns and relationships that influence loan approval decisions. Different machine learning algorithms are used to train and test the model in order to achieve high accuracy and reliable performance.

The final goal of this project is to create a smart prediction system that can assist banks in making data-driven, fair, and consistent loan approval decisions while reducing manual workload and improving efficiency.

#**Dataset Description**

The dataset used in this project is a logical and realistic loan dataset created to simulate real-world banking scenarios. It contains a large number of records representing different loan applicants, along with their financial, personal, and loan-related details.

The dataset consists of 30,000 entries (rows) and 14 attributes (columns). Each row represents a unique loan applicant, and each column represents a specific feature used to evaluate loan eligibility.

🔹 Features in the Dataset

Age – Age of the applicant (in years), used to check eligibility criteria

Gender – Male or Female

Marital Status – Whether the applicant is married or not

Education – Applicant’s education level (Graduate / Non-Graduate)

Employment Type – Type of employment (Salaried / Self-Employed)

Annual Income – Total yearly income of the applicant

Loan Amount – Amount of loan requested

Loan Term – Duration of the loan (in months/years)

Credit Score – Numerical score indicating creditworthiness

Existing Loans – Number of loans already taken by the applicant

Debt-to-Income Ratio (DTI) – Ratio of total debt to income

Savings Amount – Total savings of the applicant

Collateral Value – Value of assets pledged against the loan

Loan Status (Target Variable) – Final decision (Approved / Rejected)

🔹 Dataset Characteristics

Contains both numerical and categorical data

Designed to be balanced and realistic for training ML models

Includes financial indicators that directly impact loan decisions

Suitable for classification problems in machine learning

```python
import pandas as pd
import numpy as np

def get_int(msg):
    while True:
        try:
            return int(input(msg))
        except ValueError:
            print("Enter valid integer")

def get_float(msg):
    while True:
        try:
            return float(input(msg))
        except ValueError:
            print("Enter valid number")

def get_str(msg):
    return input(msg).strip()

def live_prediction(pipeline):

    print("\n====== LOAN APPROVAL PREDICTION SYSTEM ======\n")

    # Get raw user inputs
    age = get_int("Enter Age: ")
    gender_str = get_str("Enter Gender (Male/Female): ")
    marital_status_str = get_str("Enter Marital Status (Single/Married): ")
    education_str = get_str("Enter Education (Not Graduate/High School/Graduate/Post Graduate/Bachelor/Master/PhD): ")
    employment_type_str = get_str("Enter Employment Type (Salaried/Student/Self-Employed/Part-Time/Unemployed): ")
    annual_income = get_float("Enter Annual Income: ")
    loan_amount = get_float("Enter Loan Amount: ")
    loan_term = get_int("Enter Loan Term (in months): ")
    credit_score = get_int("Enter Credit Score: ")
    existing_loans = get_int("Enter Number of Existing Loans: ")
    dti_ratio = get_float("Enter DTI Ratio: ")
    savings_amount = get_float("Enter Savings Amount: ")
    collateral_value = get_float("Enter Collateral Value: ")

    # Create a DataFrame from user inputs
    user_data_df = pd.DataFrame([[age, gender_str, marital_status_str, education_str, employment_type_str,
                                  annual_income, loan_amount, loan_term, credit_score, existing_loans,
                                  dti_ratio, savings_amount, collateral_value]],
                                columns=['Age', 'Gender', 'Marital_Status', 'Education', 'Employment_Type',
                                         'Annual_Income', 'Loan_Amount', 'Loan_Term', 'Credit_Score',
                                         'Existing_Loans', 'DTI_Ratio', 'Savings_Amount', 'Collateral_Value'])

    # Apply the same preprocessing steps as training data

    # 1. Gender mapping
    user_data_df['Gender'] = user_data_df['Gender'].map({'Male': 1, 'Female': 0})

    # 2. Marital_Status One-Hot Encoding
    # These categories must match those generated during training data preprocessing
    marital_status_categories = ['Married', 'Single']
    for cat in marital_status_categories:
        user_data_df[f'Marital_Status_{cat}'] = (user_data_df['Marital_Status'].str.lower() == cat.lower()).astype(int)
    user_data_df = user_data_df.drop('Marital_Status', axis=1)

    # 3. Education Ordinal Encoding
    education_map = {
        'Not Graduate': 0,
        'High School': 1,
        'Graduate': 2,
        'Post Graduate': 3,
        'Bachelor': 4,
        'Master': 5,
        'PhD': 6
    }
    # Convert input to lowercase for case-insensitive matching
    user_data_df['Education'] = user_data_df['Education'].str.title().map(education_map).fillna(-1) # Use -1 for unknown types

    # 4. Employment_Type One-Hot Encoding
    # These categories must match those generated during training data preprocessing
    employment_type_categories = ['Part-Time', 'Salaried', 'Self-Employed', 'Student', 'Unemployed']
    for cat in employment_type_categories:
        user_data_df[f'Employment_Type_{cat}'] = (user_data_df['Employment_Type'].str.lower() == cat.lower()).astype(int)
    user_data_df = user_data_df.drop('Employment_Type', axis=1)

    # Define the exact order of columns expected by the trained model (rf_model)
    # This list comes from the X DataFrame used for training.
    # Check the X variable in kernel state for the correct list if unsure.
    expected_columns = ['Age', 'Gender', 'Education', 'Annual_Income', 'Loan_Amount', 'Loan_Term',
                        'Credit_Score', 'Existing_Loans', 'DTI_Ratio', 'Savings_Amount',
                        'Collateral_Value', 'Marital_Status_Married', 'Marital_Status_Single',
                        'Employment_Type_Part-Time', 'Employment_Type_Salaried',
                        'Employment_Type_Self-Employed', 'Employment_Type_Student',
                        'Employment_Type_Unemployed']

    # Ensure the columns in user_data_df match the expected order and count
    # If any expected column is missing (e.g., if a category never appeared in input),
    # add it with 0. Also handles cases where user input for a one-hot category is absent.
    for col in expected_columns:
        if col not in user_data_df.columns:
            user_data_df[col] = 0

    # Reorder columns to match the training data
    input_data = user_data_df[expected_columns]

    # Prediction
    prediction = pipeline.predict(input_data)[0]

    # Output
    if prediction == 1:
        print("\n✅ Loan Approved")
    else:
        print("\n❌ Loan Rejected")

live_prediction(rf_model)
```
